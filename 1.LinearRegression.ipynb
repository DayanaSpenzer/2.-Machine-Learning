```python
#pandas library is for manipulating data specially files and tabular data
import pandas as pd
```

```python
#reading the file Salary_Data.csv and loading the data into the variable dataset
dataset=pd.read_csv("Salary_Data.csv")
#printing dataset variable to see how the data is loaded into dataset variable
dataset
```

```python
#The next step is to split input and output data
```

```python
#fetching a column(YearsExperience) of data and loading into an independent variable
independent=dataset[["YearsExperience"]]
#checking how the data is loaded into independent variable
independent
```

```python
#fetching a column(Salary) of data and loading into a dependent variable
dependent=dataset[["Salary"]]
#checking how the data is loaded into dependent variable
dependent
```

```python
#The next step is to split test and train data set
```

```python
#imprting sklearn library to use machine learning functions
#we are importing model_selection module from sklearn and using the function train_test_split 
#to split random train and test datasets
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python
#Model Creation Phase
#Actual Learning happens in this phase 
#Slope and intercept is being calculated in the below step
```

```python
from sklearn.linear_model import LinearRegression
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#printing the slope value
weight=regressor.coef_
weight
```

```python
#printing the intercept value
bias=regressor.intercept_
bias
```

```python
#Find the predicted output with the model created
y_pred=regressor.predict(x_test)
```

```python
#Metric Evaluation Phase
#Finding R2 Value. If R2 is nearly 0 then the model created is poor and if R2 is nearly 1 then the model is good
```

```python
from sklearn.metrics import r2_score
r_square=r2_score(y_test,y_pred)
r_square
```
